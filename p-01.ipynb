```python
q1=np.percentile(loan_ds["ApplicantIncome"],25)
q2=np.percentile(loan_ds["ApplicantIncome"],50)
q3=np.percentile(loan_ds["ApplicantIncome"],75)
iqr=q3-q1
upper_bound=q3+1.5*(iqr)
lower_bound=q1-1.5*(iqr)
a=loan_ds["ApplicantIncome"]<upper_bound
b=loan_ds["ApplicantIncome"]>lower_bound
k=loan_ds["ApplicantIncome"][a & b]
pd.DataFrame(k)
```

**so we remove the oultiers as values which are greater than upperbound(q3+) and lower bound(q1-)**


- **method-2**
- **let us try by imputing the median value**
- **in this case we are replacing the outliers values with the median values**

```python
l1=[]
q1=np.percentile(loan_ds["ApplicantIncome"],25)
q2=np.percentile(loan_ds["ApplicantIncome"],50)
q3=np.percentile(loan_ds["ApplicantIncome"],75)
iqr=q3-q1
upper_bound=q3+1.5*(iqr)
lower_bound=q1-1.5*(iqr)
for i in loan_ds["ApplicantIncome"]:
    if i>upper_bound or i<lower_bound:
        l1.append(q2)
    else:
        l1.append(i)
ApI=pd.DataFrame(l1,columns=["ApplicantIncome"])
ApI
```
